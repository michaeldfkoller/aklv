```python
import numpy as np
import matplotlib.pyplot as plt

def qx(x):
    value = min(0,-7.85785 + 0.01538*x + 5.77355 * 0.0001*x*x)
    qxvalue = np.exp(value)
    return(value, qxvalue)


x = range(120)
v=[]
vqx=[]

for i in x:
    n1,n2 = qx(i)
    v.append(n1)
    vqx.append(n2)

plt.figure(1)
plt.subplot(2,1,1)
plt.plot(x,v)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,vqx)
plt.grid(True)

    
```
